```python
def compare_scd_cer(
    res, systems, chain, title, save=False, ymax=0.5,
    get_name_folder='data/outer1/151/allscd'
):
    to_return = []
    colors = ['r', 'b', 'orange', 'g', 'c', 'k']
    xs, fnames = get_x_file_cer(res, get_name_folder)
    c2s, c3s, dic = sort_names_cer(xs, fnames)
    plt.figure(figsize=(6, 4))
    plt.rc('font', size=14)
    for i, sys in enumerate(systems):
        if 'outer' in sys:
            folder = './data/{}/*/allscd/'.format(sys)
        else:
            folder1 = './data/{}*/*/scd_up/'.format(sys)
            folder2 = './data/{}*/scd_up/'.format(sys)
        c2s_scd = []
        c2s_err = []
        c3s_scd = []
        c3s_err = []
        for c in c2s:
            data = []
            try:
                files = glob.glob(folder + '{}.scd'.format(dic[c]))
            except:
                files = glob.glob(folder1 + '{}.scd'.format(dic[c])) + \
                glob.glob(folder2 + '{}.scd'.format(dic[c]))
            for f in files:
                data.append(np.loadtxt(f))
            data = np.array(data)
            mean = data.mean()
            err = data.std() / np.sqrt(data.shape[0])
            c2s_scd.append(np.abs(mean))
            c2s_err.append(err)
        for c in c3s:
            data = []
            try:
                files = glob.glob(folder + '{}.scd'.format(dic[c]))
            except:
                files = glob.glob(folder1 + '{}.scd'.format(dic[c])) + \
                glob.glob(folder2 + '{}.scd'.format(dic[c]))
            for f in files:
                data.append(np.loadtxt(f))
            data = np.array(data)
            mean = data.mean()
            err = data.std() / np.sqrt(data.shape[0])
            c3s_scd.append(np.abs(mean))
            c3s_err.append(err)
        if chain == 2:
            plt.errorbar(
                np.arange(0, len(c2s)), c2s_scd, yerr=c2s_err, 
                fmt='s', label=sys, fillstyle='none', color=colors[i]
            )
            ticks = []
            for c in c2s:
                ticks.append(c[1:-1])
            plt.xticks(np.arange(0, len(c2s)), ticks, fontsize=20, rotation=90)
        to_return.append(c2s_scd)
        if chain == 3:
            plt.errorbar(
                np.arange(0, len(c3s)), c3s_scd, yerr=c3s_err, 
                fmt='s', label=sys, fillstyle='none', color=colors[i]
            )
            ticks = []
            for c in c3s:
                ticks.append(c[1:-1])
            plt.xticks(np.arange(0, len(c3s)), ticks, fontsize=20, rotation=90)
    plt.legend(loc=(1.05, 0.0), fontsize=18)
    plt.yticks(np.arange(0, 0.51, 0.1), fontsize=20)
    plt.title(title, pad=-17, fontsize=18)
    plt.ylim(-0.03, ymax)
    if save:
        plt.savefig('{}.png'.format(save), dpi=330, bbox_inches='tight')
    plt.show()
    plt.close()
    return to_return
```

```python
def pearson(v1, v2):
    """
    v1, v2 should be 1d array
    """
    mean1 = np.mean(v1)
    mean2 = np.mean(v2)
    return np.sum((v1-mean1) * (v2-mean2)) / np.sqrt(np.sum((v1-mean1)**2) * np.sum((v2-mean2)**2))
```
